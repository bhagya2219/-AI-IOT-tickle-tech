```python
#step-3:handling missing data
#scenario-1:missing values in numerical column
#calculate mean or average of that column
#scenario-2:missing values in non-numerical column
#Fill the empty value highly repeted in that column
#scenario-3:having missing values in huge dataset
#example:20000 dataset two rows have empty values out of 20000
#remove those empty values rows

import numpy as np
from sklearn.impute import SimpleImputer #

imputer= SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
```

```python
X
```

```python
from math import remainder
#step-4:handling the catagorial data
#which converts the extra data into numeric data.
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')#encoder is a method which converts text to numeric
X=ct.fit_transform(X)
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
Y
```

```python
#step5:spilt the data into training and testing data
from sklearn.model_selection import train_test_split#train test split is a class which was imported from sk. model_selection #it is used to spliting the data into training and testing
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#step6:feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train[:,3:]=sc.fit_transform(X_train[:,3:])
X_test[:,3:]=sc.transform(X_test[:,3:])
```

```python
#machine learning
#1.supervisioned learning-when we have labeled data,and the model tasks are classification or prediction
#a.linear regression
#b.logistic regression
#2.un-supervisioned learning-unlabeled data,model remembers the patterns,we can perform prediction also.
#3.reinforcement learning-model will learn from failures(trail and error method)
```

```python
#logistic regression algorithm-when model perform a classification task we have to apply logistic regression alogorithm
#linear regression algorithm-when model performs prediction task we have to apply linear regression algorithm
```

```python
#step7:apply algorithm to the dataset
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
```
